```python
def EDA(df):
    df['거래일자'] = df['년'].astype(str) +'-'+ df['월'].astype(str) +'-'+ df['일'].astype(str)
    df['거래일자'] = pd.to_datetime(df['거래일자'], format=('%Y-%m-%d'))
    df = pd.merge(df,df_loc_sgg)
    
    return df
```

```python
df_loc = pd.read_csv('./data/법정동코드.csv', encoding='euc-kr')
df_loc = df_loc[df_loc['폐지여부'] == '존재']
df_loc['지역코드'] = df_loc['법정동코드'].astype(str).str[:5]
df_loc['읍면동코드'] = df_loc['법정동코드'].astype(str).str[5:]
```

```python
df_loc_sgg = df_loc[df_loc['읍면동코드'] == '00000']
df_loc_sgg = df_loc_sgg[df_loc_sgg['지역코드'].str[2:] != '000']
df_loc_sgg.reset_index(drop='index', inplace=True)
```

```python
from PublicDataReader import TransactionPrice

service_key = "m%2BPqC7Ztq1Mtd%2B5Cg8TgynTKW2%2BI%2FLArkK2UkYbhmQSwc3GAS4FFMcRZGtdP7rs%2Fw4HKZNiIzNVvx5w8lR%2FiIQ%3D%3D"
api = TransactionPrice(service_key)
```

```python
#280 할 차례
i= 260
df_price = pd.DataFrame([])
s_year_month = '202203'
e_year_month = str(datetime.today().year)+str(datetime.today().month).zfill(2)
for sgg in df_loc_sgg['지역코드'][i-20:i]:
    df_sold = api.get_data(
        property_type="아파트",
        trade_type="매매",
        sigungu_code = sgg,
        start_year_month = s_year_month,
        end_year_month = e_year_month
        )
    df_price = pd.concat([df_price, df_sold])
df_price = EDA(df_price)
df_price.to_csv(f'./extract_data/df_price_{i}.csv')
df_price['법정동명'].unique()
```
