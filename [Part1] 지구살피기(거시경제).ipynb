```python
start_date = '2020-01-01'
start_date_obj = datetime.strptime(start_date, '%Y-%m-%d')
start_year = start_date_obj.year

# Calculate the date two months ago
three_months_ago = now - relativedelta(months=3)
```

## 1. 장단기 금리차

```python
plt.figure(figsize=(8, 3))
sns.lineplot(x='Date', y='장단기금리차(10Y-2Y)[%]', data=T10Y2Y[T10Y2Y['Date'] >= start_date], color='blue')

# Add a gray bar between the two dates
plt.axvspan(three_months_ago,now,
            facecolor='gray', alpha=0.2)

plt.text(three_months_ago, 0.5, '최근3개월' )
plt.axhline(0, color='red', linestyle=':')
plt.title(f'장단기금리차 추이 ({start_year}년 이후)', fontsize=12)

plt.show()
```

### 4. 일본 환율/도요타

```python
plt.figure(figsize=(8, 3))

# First line plot with the primary y-axis
ax1 = sns.lineplot(x='Date', y='Yen[￥] to USD[＄]', data=DEXJPUS[DEXJPUS['Date'] >= start_date]
                   , color='blue', label='Yen[￥] to USD[＄]')
ax1.set_ylabel('Yen[￥] to USD[＄]')

# Second line plot with the secondary y-axis
ax2 = ax1.twinx()
sns.lineplot(x='Date', y='Close', data=TYO[TYO['Date'] >= start_date]
             , color='green', ax=ax2, label='TYO[＄]')
ax2.set_ylabel('Close[$]')

# Add a gray bar between the two dates
plt.axvspan(three_months_ago, now, facecolor='gray', alpha=0.2)

# Add horizontal lines with specific values
ax1.axhline(80, color='red', linestyle=':')
ax1.axhline(105, color='red', linestyle=':')
ax1.axhline(125, color='red', linestyle=':')

# Add a label for the gray bar
ax1.text(three_months_ago, 110, '최근3개월')

# Create handles and labels
handles, labels = [(a + b) for a, b in zip(ax1.get_legend_handles_labels(), ax2.get_legend_handles_labels())]
# Move the legend to the top left corner and use the custom handles and labels
plt.legend(handles, labels, loc='upper left')

plt.title(f'엔화 - 도요타 주가 추이 ({start_year}년 이후)', fontsize=12)
plt.show()
```
